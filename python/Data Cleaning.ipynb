### Data Handling Confirmed

```python
import pandas as pd
```

```python
df_confirmed = pd.read_csv('../data/data_Deaths.csv')
df_confirmed.head()
```

```python
df = df_confirmed.transpose()
df.columns = df.iloc[0]
df = df.drop("Country/Region") 
df = df.sort_index()
df.reset_index(inplace=True, drop=False)
df = df.rename(columns={'index': 'Date'})
df['Date'] = pd.to_datetime(df['Date'], format="%Y%m%d")
df
```

```python
df.to_csv("newData/Death.csv", index=False)
```

```python
df.columns
```

```python
df_new = df[['Date']]
df_new[df.columns[1:]] = df[df.columns[1:]].diff()
df_new
```

```python
df_new.iloc[0] = df.iloc[0] 
num = df_new._get_numeric_data()
num[num < 0] = 0
df_new
```

```python
df_new.to_csv("newData/Death_new.csv", index=False)
```
